```python
def fib():
    val1, val2 = 0, 1
    yield val1
    yield val2
    
    while True:
        val1, val2 = val2, val1 + val2
        yield val2
```

```python
seq = fib()
for i in range(20):
    print(next(seq))
```

```python
def measure(func, *args, **kwargs):
    import time
    def count_time(*args, **kwargs):
        start = time.time()
        out = func(*args, **kwargs)
        stop = time.time()
        print(stop - start)
        return out
    return count_time
```

#### тестирование

```python
@measure
def print_fib(x):
    seq = fib()
    for i in range(x):
        next(seq)


print_fib(100000)
```

```python
@measure
def shuffle_dict(oriented_list):
    # случайным образом перемешивает значения для ключей в словаре
    import random
    result = {}
    keys = oriented_list.keys()
    values = list(oriented_list.values())
    random.shuffle(values)
    
    for key, val in zip(keys, values):
        result[key] = val
    return result


seq1 = fib()
dict1 = {i: next(seq1) for i in range(1, 501)}

shuffle_dict(dict1)
```

```python
def flatten(two_dimensional_list):
    return [x for lst in two_dimensional_list for x in lst]
```

```python
flatten([[2, 8, 16, 'ad'], [14], [True, False]])
```
